```python
#ticket,cabin,passengerid,embarked
```

```python
titanic.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
titanic.info()
```

```python
sb.heatmap(titanic.isnull())
```

```python
def myage(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 30
        else:
            return 25
    else:
        return Age
```

```python
titanic['Age']=titanic[['Age','Pclass']].apply(myage,axis=1)
```

```python
sb.heatmap(titanic.isnull())
```

```python
#data is cleaned now
```

```python
titanic[['Age','Pclass']]
```

```python
titanic.info()
```

```python
gender=pd.get_dummies(titanic['Sex'],drop_first=True)
```

```python
gender
```

```python
embark=pd.get_dummies(titanic['Embarked'],drop_first=True)
```

```python
titanic.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
titanic.columns
```

```python
titanic=pd.concat([titanic,gender,embark],axis=1)
```

```python
titanic.columns
```

```python
#above data is ready for machin learning
```

```python
X=titanic[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'male', 'Q',
       'S']]
y=titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
aitm=LogisticRegression()
```

```python
aitm.fit(X_train,y_train)
```

```python
pred=aitm.predict(X_test)
```

```python
pred
```

```python
plt.scatter(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
```

```python
accuracy_score(y_test,pred)
```

```python
f1_score(y_test,pred)
```
